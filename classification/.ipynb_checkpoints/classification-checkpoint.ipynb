```python
# Make price binary
mean = df["price"].mean()
```

```python
df["above_average"] = (df["price"] >= mean).values.astype(int)
```

```python
df
```

```python
data = df.copy()
data = data.drop(columns=["price"])
```

```python
# Split the dataset

df_training, df_test = train_test_split(data, test_size=0.2, random_state=42)
df_train , df_val = train_test_split(df_training, test_size=0.2, random_state=42)
```

## Question 3

```python
df_train.reset_index(drop=True)
df_val.reset_index(drop=True)
df_test.reset_index(drop=True)
```

```python
y_train = df_train.above_average.values
y_val = df_val.above_average.values
y_test = df_test.above_average.values
```

```python
# We use the training set only
df_categorical = df_train.copy()
df_categorical = df_categorical.drop(columns=["Year", "Engine HP" , "Engine Cylinders", "highway MPG", "city mpg", "above_average"])
```

```python
df_categorical.info()
```

```python
def calculate_mutual_information(series):
    return mutual_info_score(series, df_train.above_average)
```

```python
df_mi = df_categorical.apply(calculate_mutual_information)
df_mi = df_mi.sort_values(ascending=False).to_frame(name='MI')
df_mi
```

#### The variable who has the lowest mutual information score is "Transmission Type". 

## Question 4

```python
df_train = df_train.drop(columns=["above_average"])
df_val = df_val.drop(columns=["above_average"])
df_test = df_test.drop(columns=["above_average"])
```

```python
#One-Hot Encoding with DictVectorizer()
train_dict = df_train.to_dict(orient = "records")
dict_vectorizer = DictVectorizer(sparse=False)
X = dict_vectorizer.fit_transform(train_dict)
```

```python
# Fit the model on the training dataset
model = LogisticRegression(solver="liblinear", C=10, max_iter=1000, random_state=42)
model.fit(X, y_train)
```

```python
# The model prediction
val_dict = df_val.to_dict(orient = "records")
X_val = dict_vectorizer.transform(val_dict)
y_pred = model.predict(X_val)
```

```python
# Accuracy on the validation dataset
accuracy = accuracy_score(y_val, y_pred)
print(accuracy.round(2))
```

#### The accuracy on the validation dataset is 0.95.

## Question 5

```python
# Feature elimination technique
features = ["Year", "Engine HP" , "Engine Cylinders", "highway MPG", "city mpg", 
           "Make", "Model", "Transmission Type", "Vehicle Style"]
```

```python
for i in features:
    features_subset = features.copy()
    features_subset.remove(i)
    
    features_dict = df_train[features_subset].to_dict(orient = "records")
    
    dict_vectorizer = DictVectorizer(sparse=False)
    X_dict = dict_vectorizer.fit_transform(features_dict)
    
    model = LogisticRegression(solver="liblinear", C=10, max_iter=1000, random_state=42)
    model.fit(X_dict, y_train)
    
    val_dict = df_val[features_subset].to_dict(orient = "records")
    X_val = dict_vectorizer.transform(val_dict)
    
    y_pred = model.predict(X_val)
    
    score = accuracy_score(y_val , y_pred)
    print( i , accuracy - score)
```

#### The feature who has the smallest difference is Year.

## Question 6

```python
df["price"] = np.log1p(df["price"])
```

```python
# Preparing data
df_training, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train , df_val = train_test_split(df_training, test_size=0.2, random_state=42)
```

```python
df_train.reset_index(drop=True)
df_val.reset_index(drop=True)
df_test.reset_index(drop=True)

y_train = df_train.above_average.values
y_val = df_val.above_average.values
y_test = df_test.above_average.values
```

```python
df_train = df_train.drop(columns=["price"])
df_val = df_val.drop(columns=["price"])
df_test = df_test.drop(columns=["price"])
```

```python
#One-Hot Encoding with DictVectorizer()
train_dict = df_train.to_dict(orient = "records")
dict_vectorizer = DictVectorizer(sparse=False)
X = dict_vectorizer.fit_transform(train_dict)
```

```python
val_dict = df_val.to_dict(orient = "records")
X_val = dict_vectorizer.transform(val_dict)
```

```python
# Ridge regression
for a in [0, 0.01, 0.1, 1, 10]:
    model = Ridge(alpha=a, solver="sag", random_state=42)
    model.fit(X, y_train)
    
    y_pred = model.predict(X_val)
    
    score = np.sqrt(mean_squared_error(y_val, y_pred))
    
    print(a, " : ", score.round(3))
```

#### The best RMSE on the validation set with the alphas leads is 0.
